```python
import pandas as pd
import numpy as np
from sklearn import linear_model, model_selection
from sklearn.preprocessing import LabelEncoder

# Load the data
data = pd.read_csv("insurancedataset.csv")

# Initialize the label encoder
label_encoder = LabelEncoder()

# Encode the 'sex' column
x = label_encoder.fit_transform(data['sex'])

# Prepare the features and target variable
x = np.array(x).reshape(-1, 1)
y = np.array(data.age).reshape(-1, 1)

# Split the dataset into training and testing sets
xtrain, xtest, ytrain, ytest = model_selection.train_test_split(x, y, test_size=0.3, random_state=42)

# Initialize the linear regression model
linear_regression_model = linear_model.LinearRegression()

# Train the model
linear_regression_model.fit(xtrain, ytrain)

# Test the model
ytest_predicted = linear_regression_model.predict(xtest)

# Print the actual and predicted values
print("Actual ages (test set):", ytest)
print("Predicted ages (test set):", ytest_predicted)
```

```python
from sklearn import linear_model, model_selection, preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Load the dataset
data=pd.read_csv("insurance.csv")
#Data Preprocessing: Label Encoding
label_encoder=preprocessing.LabelEncoder()
x=np.array(label_encoder.fit_transform(data.sex)).reshape(-1,1)
y=np.array(data.expenses).reshape(-1,1)
#split the dataset
xtrain, xtest, ytrain, ytest=model_selection.train_test_split(x,y,test_size=0.3)
#Load the model
lr_model=linear_model.LinearRegression()
lr_model.fit(xtrain, ytrain)
#test the model
#xtest=np.array([1,4.3,1.1]).reshape(-1,1)
ytest=lr_model.predict(xtest)
#Visualize the model
plt.scatter(xtrain, ytrain, color="red")
plt.scatter(xtest, ytest, color="blue")
plt.show()
```
