```python
import pandas as pd
import numpy as np
from sklearn import linear_model, model_selection
from sklearn.preprocessing import LabelEncoder

# Load the data
data = pd.read_csv("insurancedataset.csv")

# Initialize the label encoder
label_encoder = LabelEncoder()

# Encode the 'sex' column
x = label_encoder.fit_transform(data['sex'])

# Prepare the features and target variable
x = np.array(x).reshape(-1, 1)
y = np.array(data.age).reshape(-1, 1)

# Split the dataset into training and testing sets
xtrain, xtest, ytrain, ytest = model_selection.train_test_split(x, y, test_size=0.3, random_state=42)

# Initialize the linear regression model
linear_regression_model = linear_model.LinearRegression()

# Train the model
linear_regression_model.fit(xtrain, ytrain)

# Test the model
ytest_predicted = linear_regression_model.predict(xtest)

# Print the actual and predicted values
print("Actual ages (test set):", ytest)
print("Predicted ages (test set):", ytest_predicted)
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model, model_selection
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'Weather.csv'  # Update this with the correct file path
weather_data = pd.read_csv(file_path)

# Select relevant columns
data = weather_data[['MinTemp', 'Snowfall']]

# Remove non-numeric values
data = data[pd.to_numeric(data['Snowfall'], errors='coerce').notnull()]

# Convert columns to numeric
data['MinTemp'] = pd.to_numeric(data['MinTemp'], errors='coerce')
data['Snowfall'] = pd.to_numeric(data['Snowfall'], errors='coerce')

# Drop rows with any remaining NaN values
data = data.dropna()

# Data Preprocessing
X = np.array(data['MinTemp']).reshape(-1, 1)
y = np.array(data['Snowfall']).reshape(-1, 1)

# Split the dataset
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=42)

# Load the model
lr_model = linear_model.LinearRegression()
lr_model.fit(X_train, y_train)

# Generate predictions
y_pred = lr_model.predict(X_test)

# Create a range of values for plotting the regression line
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_range_pred = lr_model.predict(X_range)

# Visualize the model
plt.scatter(X_train, y_train, color="red", label="Training data")
plt.scatter(X_test, y_test, color="green", label="Test data")
plt.plot(X_range, y_range_pred, color="blue", label="Regression line")
plt.xlabel('Min Temperature')
plt.ylabel('Snowfall')
plt.legend()
plt.show()
```
