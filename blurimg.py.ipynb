```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("modi.webp")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
k1 = np.array([[0,0,0],[0,1,0],[0,0,0]])
filtered_img = cv2.filter2D(img,-1,k1)
plt.imshow(filtered_img)
```

```python
# k1 = np.array([[0,0,0],[0,1,0],[0,0,0]])
k1 = np.array([[-1,-1,-1],[-1,12,-1],[-1,-1,-1]])
filtered_img = cv2.filter2D(img,-1,k1)
plt.imshow(filtered_img)
```

```python
# blur(image,(x,y))
blur_img = cv2.blur(img,(5,5))
plt.imshow(blur_img)
```

```python
plt.imshow(img)
```

```python
image1_np = np.array(img)
image2_np = np.array(blur_img)
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(image1_np)
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(image2_np)
axs[1].set_title('Blur Image')
axs[1].axis('off') 

plt.tight_layout() 
plt.show()
```
